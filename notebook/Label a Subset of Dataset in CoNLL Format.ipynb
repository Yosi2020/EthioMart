```python
import pandas as pd
import re
import random
```

```python
INPUT_TXT    = "labeled_telegram_product_price_location_cleaned.txt"
OUTPUT_CONLL = "labeled_telegram_product_price_location_cleaned.conll"
```

```python
def normalize_amharic(text: str) -> str:
    # strip zero-width joiners, collapse whitespace
    text = re.sub(r'[\u200c\u200d]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

def tokenize_amharic(text: str) -> list[str]:
    # split off punctuation into its own tokens
    return re.findall(r'\w+|[^\s\w]', text)
```

```python
# 1) Read all non-blank lines as separate messages
with open(INPUT_TXT, 'r', encoding='utf-8') as f:
    messages = [line.strip() for line in f if line.strip()]

# 2) Write CoNLL with default O labels
with open(OUTPUT_CONLL, 'w', encoding='utf-8') as fout:
    for msg in messages:
        norm   = normalize_amharic(msg)
        tokens = tokenize_amharic(norm)
        for tok in tokens:
            fout.write(f"{tok}\tO\n")
        fout.write("\n")

print(f"Wrote {len(messages)} messages → {OUTPUT_CONLL}")
```
